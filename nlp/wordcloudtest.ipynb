### 除了直接读入文本生成词云，也可以使用字典格式的词频作为输入

```python
# 读入一组词频字典文件
text_dict = {   'you': 2993,   'and': 6625,   'in': 2767,   'was': 2525,   'the': 7845,}
wordcloud = WordCloud().generate_from_frequencies(text_dict)
# 显示词云图片
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### 还可以将词云填充到指定的形状之中。为达到填充指定形状的效果，需要使用 png 格式的图片。

```python
# 读入图片
from scipy.misc import imread
# bg_pic = imread('xiniu.png')
bg_pic = imread('yezi.png')
# 配置词云参数
wc = WordCloud(
            # 设置字体
            font_path = 'C:\\Windows\\Fonts\\STFANGSO.ttf',
            # 设置背景色
            background_color='white',
            # 允许最大词汇
            max_words=200,
            # 词云形状
            mask=bg_pic,
            # 最大号字体
            max_font_size=100,
            )
wordcloud = wc.generate(text2)
# 显示词云图片
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
wc.to_file("yezicloud.jpg")
```

### 还可以结合 jieba分词，实现用中文填充的词云图，使用中文填充时一定得指定中文的字体，否则会出现乱码。如果报错找不到字体，就复制一个中文字体文件放在代码目录下。
使用分词生成的词云更加细粒度！分词与否生成的词云各有风味！

```python
import jieba
from collections import Counter
```

```python
# 使用 jieba 分词
text_jieba = list(jieba.cut(text))

# 使用 counter 做词频统计，选取出现频率前 100 的词汇
c = Counter(text_jieba)
common_c = c.most_common(100)
common_c
```

```python
# 读入图片
from scipy.misc import imread
# bg_pic = imread('xiniu.png')
bg_pic = imread('yezi.png')
# 配置词云参数
wc = WordCloud(
            # 设置字体
            font_path = 'C:\\Windows\\Fonts\\STFANGSO.ttf',
            # 设置背景色
            background_color='white',
            # 允许最大词汇
            max_words=200,
            # 词云形状
            mask=bg_pic,
            # 最大号字体
            max_font_size=100,
            )
wordcloud = wc.generate_from_frequencies(dict(common_c))
# 显示词云图片
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
wc.to_file('jieba.jpg')
```

### 1. 编程题

根据课程中给出的1958-1969年间的数据，分别构建一元线性回归模型和倒数模型表示的菲利普斯曲线。

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('02/hw02data.csv',encoding='gbk')
df.head()
```

```python
s = df['收入指数']
```

```python
s.values
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(df['收入指数'].values[:,np.newaxis],df['失业率'].values)
```

```python
lm.coef_, lm.intercept_
```

所以，一元线性模型：y = 7.666604197901049 -0.65367316X

```python
lm1 = linear_model.LinearRegression()
```

```python
lm1.fit(1 / df['收入指数'].values[:,np.newaxis],df['失业率'].values)
```

```python
lm1.coef_, lm1.intercept_
```

```python
plt.scatter(df['收入指数'].values[:,np.newaxis],df['失业率'].values)
plt.plot(df['收入指数'].values[:,np.newaxis], lm.predict(df['收入指数'].values[:,np.newaxis]))
plt.plot(df['收入指数'].values[:,np.newaxis], lm1.predict(1/df['收入指数'].values[:,np.newaxis]),color='red')
plt.show()
```

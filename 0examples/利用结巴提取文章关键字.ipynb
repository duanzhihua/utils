```python
import sys
sys.path.append('/home/ian/code/github')
```

```python
from utils.crawl.requests_wrapper import requests_wrapper as rqw
```

```python
rqw = rqw('http://bi.dataguru.cn/article-9776-1.html')
```

```python
text = rqw.get_text(encoding='gbk')
```

```python
from utils.crawl.pyquery_wrapper import pyquery_wrapper as pqw
```

```python
doc = pqw(text)
```

```python
str1 = doc.Root('#article_content').text()
str1
```

```python
import jieba
```

```python
words = jieba.lcut(str1)
words[:5]
```

```python
words = [w for w in words if w.strip()]
words[:5]
```

```python
import nltk
```

```python
#查看最常出现的20个词
nltk.FreqDist(words).most_common(20)
```

```python
#尝试找到文章的10个关键词
import jieba.analyse
jieba.analyse.extract_tags(str1,topK=10)
```

```python
jieba.analyse.textrank(str1,topK=10)
```

```python
from pyecharts import Line
```

```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris() # 导入数据集
X = iris.data # 获得其特征向量
y = iris.target # 获得样本label
feature_names = iris.feature_names#特征名称
```

```python
line = Line("鸢尾花数据集可视化观察",title_pos='center',subtitle="练习pyecharts使用")
```

## 1、显示3种类别的鸢尾花的4种特征的变化趋势
鸢尾花数据集中提供了3种类别array(['setosa', 'versicolor', 'virginica']。  
首先，查看一下不同类别的鸢尾花之间的4个特征['sepal length (cm)',  'sepal width (cm)',  'petal length (cm)',  'petal width (cm)']有无显著差异？！

```python
line.add(
    'sepal length (cm)',
    list(range(X.shape[0])),X[:,0],
    legend_top='bottom',
    is_more_utils=True,
    mark_point=[
        'min','max',
        {'coord':[50, X[50,0]]}#显示指定的横坐标的值
    ],
    mark_point_symbolsize=80
)
line.add(
    'sepal width (cm)',
    list(range(X.shape[0])),X[:,1],
    legend_top='bottom',
    is_more_utils=True,
    mark_point=[
        'min','max',
    ],
    mark_point_symbolsize=80
)
line.add(
    'petal length (cm)',
    list(range(X.shape[0])),X[:,2],
    legend_top='bottom',
    is_more_utils=True,
    mark_point=[
        'min','max',
    ],
    mark_point_symbolsize=80
)
line.add(
    'petal width (cm)',
    list(range(X.shape[0])),X[:,3],
    legend_top='bottom',
    is_more_utils=True,
    mark_point=[
        'min','max',
    ],
    mark_point_symbolsize=80
)
line
```

## 2、对比3种类别的'petal width (cm)'对应的均值

```python
setosa_petal_width = [X[i,3] for i in range(len(y)) if y[i] == 0]

setosa_petal_width_mean = round(sum(setosa_petal_width)/len(setosa_petal_width),2)#结果保留两位小数
setosa_petal_width_mean
```

```python
versicolor_petal_width = [X[i,3] for i in range(len(y)) if y[i] == 1]

versicolor_petal_width_mean = round(sum(versicolor_petal_width)/len(versicolor_petal_width),2)
versicolor_petal_width_mean
```

```python
virginica_petal_width = [X[i,3] for i in range(len(y)) if y[i] == 2]

virginica_petal_width_mean = round(sum(virginica_petal_width)/len(virginica_petal_width),2)
virginica_petal_width_mean
```

```python
from pyecharts import Bar
bar = Bar("3种类别的'petal width (cm)'对应的均值对比",title_pos='center',subtitle="练习pyecharts使用")
bar.add('setosa_petal_width',['width'],np.array([setosa_petal_width_mean]),legend_top='bottom',mark_point=["average"],mark_point_symbol='roundRect')
bar.add('versicolor_petal_width',['width'],np.array([versicolor_petal_width_mean]),legend_top='bottom',mark_point=["average"],mark_point_symbol='roundRect')
bar.add('virginica_petal_width',['width'],np.array([virginica_petal_width_mean]),legend_top='bottom',mark_point=["average"],mark_point_symbol='roundRect')
bar
```

对比，可以看出'petal length (cm)', 'petal width (cm)'不同类别的区分度是很强的, 是非常有效的特征！

## 3、研究petal width的变化差值

```python
len(versicolor_petal_width),len(virginica_petal_width)
```

```python
dif = np.array(virginica_petal_width) - np.array(versicolor_petal_width)
dif
```

```python
differ_line = Line("petal width的变化差值变化",title_pos='center',subtitle="petal width的变化差值")
differ_line.add(
    'petal width的变化差值',np.arange(len(dif)),dif,
    mark_line=["average"],#绘制出辅助平均线
    legend_top='bottom',
    is_fill=True,
    line_opacity=0.2,
    area_opacity=0.5,
    mark_point=['min','max',
#                 {'coord':[data[data['年份']=='1980年'].index,data[data['年份']=='1980年']['男女人口差值'].values[0]]}
               ]
)
differ_line
```
